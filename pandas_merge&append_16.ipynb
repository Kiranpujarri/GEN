```python
import pandas as pd

# First DataFrame: Marks
df1 = pd.DataFrame({
    'Name': ['A', 'A', 'C', 'D'],
    'Marks': [10, 20, 30, 40]
})

# Second DataFrame: Gender
df2 = pd.DataFrame({
    'Name': ['A', 'A', 'A', 'E', 'F'],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male']
})

df3 =pd.DataFrame({
    'result':['pass','fail','pass','pass','fail']
})

```

```python
df1
```

```python
pd.merge(df1,df2,on='Name')
pd.merge(df2,df3,on='result')
```

```python
pd.merge(df1,df2,on='Name')
```

```python
result = pd.merge(df1, df2, on='Name')
```

```python
import pandas as pd

# Sales Transactions Table
df_sales = pd.DataFrame({
    'CustomerID': ['A', 'A', 'C', 'D'],
    'TransactionAmount': [10, 20, 30, 40],
    'TransactionID': ['T001', 'T002', 'T003', 'T004'],
    'Date': ['2024-06-01', '2024-06-05', '2024-06-07', '2024-06-08']
})
```

```python
# Customer Details Table
df_customers = pd.DataFrame({
    'CustomerID': ['A', 'A', 'B', 'D', 'D'],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
    'Location': ['Mumbai', 'Mumbai', 'Delhi', 'Chennai', 'Hyderabad']
})
```

```python
pd.merge(df1,df2,on='Name',how='left')
```

```python
pd.merge(df2,df1,on='Name',how='right')
```

```python
pd.merge(df2,df1,on='Name',how='left',indicator=True)
```

```python
merged = pd.merge(df_sales, df_customers, on='CustomerID', how='inner')

# Select only the desired columns
result = merged[['CustomerID','Gender', 'Location', 'Date']]
```
